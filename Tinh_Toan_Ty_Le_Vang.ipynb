```python
from tkinter import *
try:                        # In order to be able to import tkinter for
    import tkinter as tk    # either in python 2 or in python 3
except:
    import Tkinter as tk
from PIL import Image, ImageTk
import cv2 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from tensorflow.keras.models import load_model
import numpy as np
from tensorflow.keras.utils import load_img, img_to_array
from tensorflow.keras.models import Sequential
from keras.models import load_model
font=cv2.FONT_HERSHEY_COMPLEX

model = load_model('ND_cnn.h5')
win = Tk()
win.title('Tỷ Lệ Vàng Trên Khuôn Mặt')
win.geometry('500x500')
def click():
    win.destroy()
    win1=Tk()
    win1.title('Tính Toán Tỷ Lệ Vàng')
    win1.geometry('1000x500')
    label = tk.Label(win1)
    filename='pexels-henry-&-co-2341290.jpg'
    img = Image.open(filename)
    label.img = ImageTk.PhotoImage(img)
    label['image'] = label.img
    label.pack()
    label.place(x=0,y=0)
    doc1=Label(win1, text='BẢNG THÔNG SỐ TÍNH TOÁN TỶ LỆ',font=('Times New Roman',16),bg='yellow')
    doc1.place(x=450,y=250)
    kcn1=Label(win1, text='Mép tai 1– đuôi mắt 1:0', font=('Times New Roman',14))
    kcn1.place(x=450,y=300)
    kcn2=Label(win1, text='Đuôi mắt 1 – cánh mũi 1:0', font=('Times New Roman',14))
    kcn2.place(x=450,y=330)
    kcn3=Label(win1, text='Cánh mũi 1 – cánh mũi 2:0', font=('Times New Roman',14))
    kcn3.place(x=450,y=360)
    kcn4=Label(win1, text='Cánh mũi 2 – đuôi mắt 2:0', font=('Times New Roman',14))
    kcn4.place(x=450,y=390)
    kcn5=Label(win1, text='Đuôi mắt 2 – mép tai 2:0', font=('Times New Roman',14))
    kcn5.place(x=450,y=420)
    kcd1=Label(win1, text='Mép tóc trên – chân mày:0', font=('Times New Roman',14))
    kcd1.place(x=750,y=300)
    kcd2=Label(win1, text='Chân mày – nhân trung:0', font=('Times New Roman',14))
    kcd2.place(x=750,y=330)
    kcd3=Label(win1, text='Nhân trung – cằm dưới:0', font=('Times New Roman',14))
    kcd3.place(x=750,y=360)
    doc=Label(win1, text='TIÊU CHUẨN TỶ LỆ VÀNG DỰA VÀO Y HỌC VIỆT NAM',font=('Times New Roman',16),bg='yellow')
    doc.place(x=450,y=20)
    doc2=Label(win1, text='-Tỷ lệ dọc 1/5: Chúng ta sẽ được 5 khoảng cách lần lượt là: Mép tai 1–    \n đuôi mắt 1; đuôi mắt 1 – cánh mũi 1; cánh mũi 1 – cánh mũi 2; cánh mũi\n 2 – đuôi mắt 2; đuôi mắt 2 – mép tai 2. Nếu cả 5 đoạn thẳng trên có       \nchiều dài bằng nhau và bằng 1/5 tổng chiều dài hai mép tai sẽ được gọi là\ntỷ lệ chuẩn theo chiều dọc. Ngoài 1:1:1:1:1 còn có tỷ lệ 0.8:1:1.2:0.8:1    ',font=('Times New Roman',14))
    doc2.place(x=450,y=50)
    doc2=Label(win1, text='-Tỷ lệ ngang 1/3: Cách chia vùng sẽ là: Mép tóc trên – chân mày; chân    \nmày – nhân trung; nhân trung – cằm dưới. Nếu cả 3 khoảng cách trên có\n độ dài bằng nhau và bằng 1/3 chiều dài từ mép tóc tới cằm dưới, bạn đã  \n đạt độ chuẩn về chiều ngang.\t\t\t\t         ',font=('Times New Roman',14))
    doc2.place(x=450,y=160)
    def click1():
        video=cv2.VideoCapture(0)
        face=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
        eye_left=cv2.CascadeClassifier('haarcascade_lefteye_2splits.xml')
        eye_right=cv2.CascadeClassifier('haarcascade_righteye_2splits.xml')
        nose=cv2.CascadeClassifier('haarcascade_mcs_nose.xml')
        mieng=cv2.CascadeClassifier('mieng.xml')
        dem=0
        while True:
            x=0;y=0;w=0;h=0
            ret,frame=video.read()
            faces=face.detectMultiScale(frame,1.3,5)
            for x,y,w,h in faces:
                cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)
                eye_rights=eye_right.detectMultiScale(frame,1.3,4)
                noses=nose.detectMultiScale(frame,1.3,8)
                eye_lefts=eye_left.detectMultiScale(frame,1.3,4)
                miengs=mieng.detectMultiScale(frame,1.3,5)
                for nose_x,nose_y,nose_w,nose_h in noses:
                    crop_img=frame[nose_y:nose_y+nose_h,nose_x:nose_x+nose_w]
                    name='C:/Users/Son-14/Desktop/AI/Nhan_Dien/'+'mui.jpg'
                    cv2.imwrite(name, crop_img)
                    img = load_img(name,target_size=(150,150))
                    img = img_to_array(img)
                    img = img.reshape(1,150,150,3)
                    img = img.astype('float32')
                    img = img/255
                    kq= np.argmax(model.predict(img),axis=-1)
                    x_mui_trai=nose_x+2
                    x_mui_phai=nose_x+nose_w
                    y_mui=nose_y+30
                    if kq==3:
                        cv2.rectangle(frame,(x_mui_trai,nose_y+30),(x_mui_trai,nose_y+30),(255, 0, 0), 3)
                        cv2.rectangle(frame,(x_mui_phai,nose_y+nose_h-25),(x_mui_phai,nose_y+nose_h-25),(255, 0, 0), 3)
                        cv2.putText(frame, 'Mui',(nose_x+20,nose_y+28), font, 0.25, (255,255,255),1, cv2.LINE_AA)
                for eye_x,eye_y,eye_w,eye_h in eye_rights:
                    crop_img=frame[eye_y:eye_y+eye_h,eye_x:eye_x+eye_w]
                    name='C:/Users/Son-14/Desktop/AI/Nhan_Dien/'+'mat_phai.jpg'
                    cv2.imwrite(name, crop_img)
                    img = load_img(name,target_size=(150,150))
                    img = img_to_array(img)
                    img = img.reshape(1,150,150,3)
                    img = img.astype('float32')
                    img = img/255
                    kq= np.argmax(model.predict(img),axis=-1)
                    if x_mui_trai>eye_x :
                        x_mat_trai=eye_x+3
                        y_mat_trai=eye_y+38
                        if kq==0:
                            cv2.rectangle(frame,(x_mat_trai,y_mat_trai),(x_mat_trai,y_mat_trai),(128, 0, 255), 3)
                            cv2.putText(frame, 'Mat_phai',(x_mat_trai,y_mat_trai-10), font, 0.25, (255,255,255),1, cv2.LINE_AA)
                for eye_x1,eye_y1,eye_w1,eye_h1 in eye_lefts:
                    crop_img=frame[eye_y1:eye_y1+eye_h1,eye_x1:eye_x1+eye_w1]
                    name='C:/Users/Son-14/Desktop/AI/Nhan_Dien/'+'mat_trai.jpg'
                    cv2.imwrite(name, crop_img)
                    img = load_img(name,target_size=(150,150))
                    img = img_to_array(img)
                    img = img.reshape(1,150,150,3)
                    img = img.astype('float32')
                    img = img/255
                    kq= np.argmax(model.predict(img),axis=-1)
                    if x_mui_phai<(eye_x1+eye_w1) :
                        x_mat_phai=eye_x1-3+eye_w1
                        y_mat_phai=eye_y1-18+eye_h1
                        if kq==1:
                            cv2.rectangle(frame,(x_mat_phai,y_mat_phai),(x_mat_phai,y_mat_phai),(128, 0, 255), 3)
                            cv2.putText(frame, 'Mat_trai',(x_mat_phai-50,y_mat_phai-15), font, 0.25, (255,255,255),1, cv2.LINE_AA)
                for lmay_x,lmay_y,lmay_w,lmay_h in eye_rights:
                        x_long_may=lmay_x
                        y_long_may=lmay_y
                        cv2.rectangle(frame,(x_long_may+20,y_long_may+5),(x_long_may+20,y_long_may+5),(0,255,0),3)
                for mieng_x,mieng_y,mieng_w,mieng_h in miengs:
                    if mieng_y>y_mui :
                        if mieng_y<y_mui+20 :
                            crop_img=frame[mieng_y:mieng_y+mieng_h,mieng_x:mieng_x+mieng_w]
                            name='C:/Users/Son-14/Desktop/AI/Nhan_Dien/'+'mieng.jpg'
                            cv2.imwrite(name, crop_img)
                            img = load_img(name,target_size=(150,150))
                            img = img_to_array(img)
                            img = img.reshape(1,150,150,3)
                            img = img.astype('float32')
                            img = img/255
                            kq= np.argmax(model.predict(img),axis=-1)
                            x_mieng=mieng_x+60
                            y_mieng=mieng_y+5
                            if kq==2:
                                cv2.rectangle(frame,(x_mieng,y_mieng),(x_mieng,y_mieng),(255,255,0),3)
                                cv2.putText(frame, 'Mieng',(x_mieng+20,y_mieng+28), font, 0.25, (255,255,255),1, cv2.LINE_AA)
            cv2.imshow("WindowCamera",frame)
            k=cv2.waitKey(1)
            # khi cắt ảnh thành 300x300 thì cắt từ tai nên là x_tai_trai và x_tai_phai sẽ là tọa độ của điểm cắt
            x_tai_trai=x+20
            x_tai_phai=x+w-12
            y_cam=y+h+5
            y_trang=y
            if k==ord('v') :
                crop = frame[y:y+h+5,x+20:x+w-12]
                crop=cv2.resize(crop, dsize=(300,300))
                name='C:/Users/Son-14/Desktop/AI/Nhan_Dien/'+'anh_customer'+str(dem)+'.jpg'
                cv2.imwrite(name, crop)
                dem=dem+1
                break
        video.release()
        cv2.destroyAllWindows()
        if dem>0:
            filename ='C:/Users/Son-14/Desktop/AI/Nhan_Dien/'+'anh_customer'+str(dem-1)+'.jpg'
            label = tk.Label(win1)
            img = Image.open(filename)
            label.img = ImageTk.PhotoImage(img)
            label['image'] = label.img
            label.pack()
            label.place(x=40,y=20)


        # print(x_tai_trai,x_mat_trai,x_mui_trai,x_mui_phai,x_mat_phai,x_tai_phai)
        chieu_ngang=(x_tai_phai-x_tai_trai)/5

        kc_n1=round(float((x_mat_trai-x_tai_trai)/chieu_ngang),2)
        kcn1=Label(win1, text='Mép tai 1– đuôi mắt 1:'+'  '+str(kc_n1)+'   ', font=('Times New Roman',14))
        kcn1.place(x=450,y=300)
    #     print("Tỷ lệ khoảng cách ngang 1:",kc_n1)

        kc_n2=round(float((x_mui_trai-x_mat_trai)/chieu_ngang),2)
        kcn2=Label(win1, text='Đuôi mắt 1 – cánh mũi 1:'+'  '+str(kc_n2)+'   ', font=('Times New Roman',14))
        kcn2.place(x=450,y=330)
    #     print("Tỷ lệ khoảng cách ngang 2:",kc_n2)

        kc_n3=round(float((x_mui_phai-x_mui_trai)/chieu_ngang),2)
        kcn3=Label(win1, text='Cánh mũi 1 – cánh mũi 2:'+'  '+str(kc_n3)+'   ', font=('Times New Roman',14))
        kcn3.place(x=450,y=360)
    #     print("Tỷ lệ khoảng cách ngang 3:",kc_n3)

        kc_n4=round(float((x_mat_phai-x_mui_phai)/chieu_ngang),2)
        kcn4=Label(win1, text='Cánh mũi 2 – đuôi mắt 2:'+'  '+str(kc_n4)+'   ', font=('Times New Roman',14))
        kcn4.place(x=450,y=390)
    #     print("Tỷ lệ khoảng cách ngang 4:",kc_n4)

        kc_n5=round(float((x_tai_phai-x_mat_phai)/chieu_ngang),2)
        kcn5=Label(win1, text='Đuôi mắt 2 – mép tai 2:'+'  '+str(kc_n5)+'   ', font=('Times New Roman',14))
        kcn5.place(x=450,y=420)
    #     print("Tỷ lệ khoảng cách ngang 5:",kc_n5)

        chieu_doc=(y_cam-y_trang)/3

        kc_d1=round(float((y_cam-y_mieng)/chieu_doc),2)
        kcd1=Label(win1, text='Mép tóc trên – chân mày:'+'  '+str(kc_d1)+'   ', font=('Times New Roman',14))
        kcd1.place(x=750,y=300)
    #     print("Tỷ lệ khoảng cách dọc 1:",kc_d1)

        kc_d2=round(float((y_mieng-y_long_may)/chieu_doc),2)
        kcd2=Label(win1, text='Chân mày – nhân trung:'+'  '+str(kc_d2)+'   ', font=('Times New Roman',14))
        kcd2.place(x=750,y=330)
    #     print("Tỷ lệ khoảng cách dọc 2:",kc_d2)

        kc_d3=round(float((y_long_may-y_trang)/chieu_doc),2)
        kcd3=Label(win1, text='Nhân trung – cằm dưới:'+'  '+str(kc_d3)+'   ', font=('Times New Roman',14))
        kcd3.place(x=750,y=360)

    but1 = Button(win1,font=('Arial',20),text='Chụp Ảnh',bg='yellow',command = click1)
    but1.place(x=100,y=360)
    win1.mainloop()
label = tk.Label(win)
filename='ty-le-mat-a-dong.jpg'
img = Image.open(filename)
label.img = ImageTk.PhotoImage(img)
label['image'] = label.img
label.pack()
label.place(x=-40,y=0)
doc=Label(win, text='TÍNH TOÁN TỶ LỆ VÀNG TRÊN KHUÔN MẶT \n DỰA VÀO Y HỌC VIỆT NAM',font=('Times New Roman',16))
doc.place(x=20,y=20)
but = Button(win,font=('Arial',20),text='VÀO',bg='yellow',command = click)
but.place(x=200,y=80)
win.mainloop()
cv2.destroyAllWindows()
```
